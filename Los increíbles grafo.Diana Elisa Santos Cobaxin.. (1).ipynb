## Distancia
Calcula todas las distancias más cortas entre pares de nodos y devuelve un diccionario donde las claves son los nodos y los valores son otros diccionarios que mapean los nodos alcanzables y sus distancias.

```python
nx.all_pairs_shortest_path_length(G) 
```

## Centro

Calcula los nodos que están en el centro del grafo, es decir, aquellos cuya máxima distancia a otros nodos es mínima.

```python
nx.center(G)
```

## Radio

Calcula el radio del grafo, que es la menor excentricidad de cualquier nodo en el grafo.

```python
nx.radius(G)
```

## Periferia

Calcula los nodos que están en la periferia del grafo, es decir, aquellos cuya excentricidad es máxima.

```python
nx.periphery(G)
```

## Excentricidad
Calcula la excentricidad de cada nodo en el grafo, que es la máxima distancia geodésica desde ese nodo a cualquier otro nodo en el grafo.

```python
nx.eccentricity(G)
```
